```python
def check_appointment_eligibility(patients):
    results = []
    eligible_count = 0
    not_eligible_count = 0

    for patient in patients:
        name = patient.get("name")
        age = patient.get("age")

        # Input validation
        if not isinstance(age, int) or age < 0:
            status = "Invalid Age"
        elif age >= 18:
            status = "Eligible"
            eligible_count += 1
        else:
            status = "Not Eligible"
            not_eligible_count += 1

        results.append({
            "name": name,
            "age": age,
            "status": status
        })

    # Display Results
    print("\n--- Appointment Eligibility Report ---")
    for record in results:
        print(f"Patient: {record['name']}")
        print(f"Age: {record['age']}")
        print(f"Eligibility Status: {record['status']}")
        print("-" * 40)

    print("Total Eligible Patients:", eligible_count)
    print("Total Not Eligible Patients:", not_eligible_count)

```

```python
# Example Usage
patient_list = [
    {"name": "Nitin", "age": 21},
    {"name": "Rahul", "age": 16},
    {"name": "Anita", "age": 30},
    {"name": "Karan", "age": -5}   # Invalid case
]

check_appointment_eligibility(patient_list)
```
