```python
def detect_duplicate_accounts(usernames):
    username_count = {}
    duplicates = {}

    # Count occurrences using loop
    for username in usernames:
        if username in username_count:
            username_count[username] += 1
        else:
            username_count[username] = 1

    # Find duplicates
    for user, count in username_count.items():
        if count > 1:
            duplicates[user] = count

    # Display results
    if duplicates:
        print("Duplicate Accounts Found: Yes")
        print("Flagged Usernames:")
        for user, count in duplicates.items():
            print(f"{user} → {count} times")
    else:
        print("Duplicate Accounts Found: No")
        print("All accounts are unique.")

    return duplicates

```

```python
# Example Usage
users = ["nitin123", "rahul_01", "techguru", "nitin123", "rahul_01"]
detect_duplicate_accounts(users)
```
