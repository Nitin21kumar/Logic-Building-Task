```python
def attendance_eligibility(attendance_list):
    total_classes = len(attendance_list)
    attended = 0

    # Count attended classes
    for status in attendance_list:
        if status == 1:   # 1 means Present
            attended += 1

    attendance_percentage = (attended / total_classes) * 100

    print("Attendance Percentage:", round(attendance_percentage, 1))

    if attendance_percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    print("Exam Eligibility:", eligibility)
    return attendance_percentage, eligibility

```

```python

attendance_data = [1, 1, 1, 0, 1]
attendance_eligibility(attendance_data)
```
