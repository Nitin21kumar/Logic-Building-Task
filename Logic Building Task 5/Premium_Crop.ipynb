```python
def premium_crop_price_filter(crops, threshold=2000):
    premium_crops = []
    regular_crops = []
    invalid_entries = []
    total_premium_value = 0

    for crop in crops:
        name = crop.get("name")
        price = crop.get("price")

        # Input validation
        if not isinstance(price, (int, float)) or price < 0:
            invalid_entries.append(name)
            continue

        if price > threshold:
            premium_crops.append(crop)
            total_premium_value += price
        else:
            regular_crops.append(crop)

    # Statistics
    avg_premium_price = (
        total_premium_value / len(premium_crops)
        if premium_crops else 0
    )

    # Display Report
    print("\n--- Premium Crop Price Report ---")

    print("\nPremium Crops:")
    for crop in premium_crops:
        print(f"{crop['name']} → ₹{crop['price']}")

    print("\nRegular Crops:")
    for crop in regular_crops:
        print(f"{crop['name']} → ₹{crop['price']}")

    if invalid_entries:
        print("\nInvalid Entries:", invalid_entries)

    print("\nTotal Premium Crops:", len(premium_crops))
    print("Average Premium Price: ₹", round(avg_premium_price, 2))

```

```python


# Example Usage
crop_data = [
    {"name": "Basmati Rice", "price": 2500},
    {"name": "Wheat", "price": 1800},
    {"name": "Almonds", "price": 3200},
    {"name": "Corn", "price": 1500},
    {"name": "Organic Tea", "price": -200}  # Invalid example
]

premium_crop_price_filter(crop_data)
```
