```python
def classify_server_load(cpu_readings):

    # Validate input list
    if len(cpu_readings) == 0:
        print("Error: No CPU readings provided.")
        return

    # Calculate total CPU usage using loop
    total_cpu = 0
    for reading in cpu_readings:
        total_cpu += reading

    # Calculate average CPU usage
    average_cpu = total_cpu / len(cpu_readings)

    print("Server Load Report")
    print(" ")
    print("Average CPU Load:", round(average_cpu), "%")
    print(" ")

    # Classify server status
    if average_cpu < 50:
        print("Server Status: Normal")

    elif 50 <= average_cpu <= 80:
        print("Server Status: Warning")

    else:
        print("Server Status: Critical")
```

```python
cpu_data = [85, 60, 90, 95, 90]
classify_server_load(cpu_data)
```
