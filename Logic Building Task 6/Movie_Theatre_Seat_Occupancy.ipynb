```python
def analyze_seat_occupancy(total_seats, seat_list):

    # Validate input length
    if len(seat_list) > total_seats:
        print("Error: Seat list exceeds total seat capacity.")
        return

    # Count booked seats (count number of 1s)
    booked_count = seat_list.count(1)

    # Calculate occupancy percentage
    occupancy_percentage = (booked_count / total_seats) * 100

    print("Theatre Occupancy Report ")
    print(" ")
    print("Occupancy:", round(occupancy_percentage), "%")
    print(" ")

    # Determine show status
    if occupancy_percentage == 100:
        print("Show Status: Housefull")

    elif occupancy_percentage >= 75:
        print("Show Status: Almost Full")

    else:
        print("Show Status: Seats Available")

    # Suggest additional show if occupancy very high
    if occupancy_percentage >= 90:
        print(" ")
        print("Suggestion: Consider Opening Additional Show")
```

```python
total_seats = 200
booked_seats = [1] * 200   # 150 seats booked

analyze_seat_occupancy(total_seats, booked_seats)
```
