```python
def manage_event_registration(event_capacity, total_registrations):

    confirmed_count = 0
    waitlist_count = 0

    # Simulate registration processing using loop
    for user in range(total_registrations):

        # If capacity not full → Confirm registration
        if confirmed_count < event_capacity:
            confirmed_count += 1

        # If capacity reached → Add to waitlist
        else:
            waitlist_count += 1

    print("Event Registration Report")
    print("Confirmed Registrations:", confirmed_count)
    print("Waitlisted Users:", waitlist_count)

    # Determine registration status
    if confirmed_count >= event_capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")
```

```python
event_capacity = 100
registrations = 105

manage_event_registration(event_capacity, registrations)
```
