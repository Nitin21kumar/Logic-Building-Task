```python
def estimate_delivery_time(distance_km, traffic_level, weather_condition):
    # Base average delivery speed in km per hour
    average_speed = 30  

    # Calculate base delivery time (convert hours to minutes)
    base_time = (distance_km / average_speed) * 60

    # Initialize delay variables
    traffic_delay = 0
    weather_delay = 0

    # Apply traffic-based delay
    if traffic_level.lower() == "low":
        traffic_delay = 5
    elif traffic_level.lower() == "medium":
        traffic_delay = 10
    elif traffic_level.lower() == "high":
        traffic_delay = 20

    # Apply weather-based delay
    if weather_condition.lower() == "clear":
        weather_delay = 0
    elif weather_condition.lower() == "rainy":
        weather_delay = 10
    elif weather_condition.lower() == "storm":
        weather_delay = 20

    # Final ETA calculation
    final_eta = base_time + traffic_delay + weather_delay

    # Display Result
    print("Delivery Time Estimation")
    print("")
    print("Estimated Delivery Time:", round(final_eta), "minutes")
    
```

```python
distance = 8
traffic = "High"
weather = "Rainy"

estimate_delivery_time(distance, traffic, weather)
```
