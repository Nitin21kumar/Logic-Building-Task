```python
def monitor_resource_usage(resource_usage):

    # Define overuse thresholds for each resource
    usage_threshold = {
        "Projector": 5,
        "AC": 8,
        "Lights": 6
    }

    overused_resources = []

    # Loop through each resource in input dictionary
    for resource, hours in resource_usage.items():

        # Check if resource exceeds its threshold
        if resource in usage_threshold and hours > usage_threshold[resource]:
            overused_resources.append(resource)

    print("Smart Classroom Energy Report")
    print(" ")

    if overused_resources:
        print("Overused Resources:", ", ".join(overused_resources))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
```

```python
resource_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resource_usage(resource_data)
```
