```python
def smart_parking_system(parking_capacity, vehicle_logs):

    # Variable to track current parked vehicles
    current_vehicles = 0

    # Variable to track highest number of vehicles parked at any time
    peak_usage = 0

    # Process each vehicle movement log
    for log in vehicle_logs:

        # If a vehicle enters parking
        if log == "IN":
            current_vehicles += 1

        # If a vehicle exits parking
        elif log == "OUT":
            # Ensure vehicle count does not go negative
            if current_vehicles > 0:
                current_vehicles -= 1

        # Update peak usage if current vehicles exceed previous peak
        if current_vehicles > peak_usage:
            peak_usage = current_vehicles

    # Display parking summary report
    print("===== Smart Parking Report =====")
    print("Currently Parked Vehicles:", current_vehicles)
    print("Peak Parking Usage:", peak_usage)

    # Check if parking exceeds maximum capacity
    if current_vehicles > parking_capacity:
        print("âš  ALERT: Parking capacity exceeded!")
    else:
        print("Parking Status: Available")
```

```python
parking_capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(parking_capacity, vehicle_logs)
```
