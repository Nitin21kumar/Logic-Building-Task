```python
# Employee ID Card System

class Employee:
    
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


```

```python
#Input
employee1 = Employee("Rahul", "EMP102", "IT")
employee2 = Employee("Raul", "EMP103", "IT")
employee3 = Employee("hulll", "EMP104", "IT")

# Display Output
employee1.display_id_card()
employee2.display_id_card()
employee3.display_id_card()
```
