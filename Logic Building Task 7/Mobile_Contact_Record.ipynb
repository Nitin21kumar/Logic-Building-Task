```python
# Mobile Contact Record System with Duplicate Check

class ContactBook:
    
    def __init__(self):
        self.contacts = {}   # Dictionary to store contacts
    
    def add_contact(self, name, phone):
        if name in self.contacts:
            print("Contact already exists")
        else:
            self.contacts[name] = phone
            print("Contact Saved")
    
    def display_contact(self, name):
        if name in self.contacts:
            print("Name:", name)
            print("Phone:", self.contacts[name])
        else:
            print("Contact not found")

```

```python
# Main Program
book = ContactBook()

# First time storing
book.add_contact("Anita", "9876543210")

# Trying to store same name again
book.add_contact("Anita", "9876543210")

# Display contact
book.display_contact("Anita")
```
