```python
# Smart Light Controller Program

class SmartLight:
    
    def __init__(self, name):
        self.light_name = name
        self.status = "OFF"   # Default status
    
    def turn_on(self):
        self.status = "ON"
    
    def turn_off(self):
        self.status = "OFF"
    
    def display_status(self):
        print(self.light_name + " is " + self.status)
```

```python
# Main Program
light = SmartLight("Bedroom Light")

light.turn_off()        # Action: ON
light.display_status() # Display result
```
